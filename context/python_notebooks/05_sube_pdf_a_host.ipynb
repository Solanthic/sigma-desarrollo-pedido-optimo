```python
import os
import paramiko

# Configuración de conexión SFTP
host = "162.241.61.143"
port = 2222
username = "sopor152"
password = "Viejo#2007"
output_folder = os.path.join('alertas_wp', 'impulso')
remote_folder = "/home1/sopor152/public_html/shared/analytics/impulso_pdf/"

# Crear conexión SFTP
transport = paramiko.Transport((host, port))
transport.connect(username=username, password=password)
sftp = paramiko.SFTPClient.from_transport(transport)

# Subir archivos PDF y sobrescribir si existen
for filename in os.listdir(output_folder):
    if filename.lower().endswith(".pdf"):
        local_path = os.path.join(output_folder, filename)
        remote_path = os.path.join(remote_folder, filename)
        try:
            sftp.put(local_path, remote_path)
            print(f"Archivo sobrescrito/subido: {filename}")
        except Exception as e:
            print(f"Error al subir {filename}: {e}")

# Cerrar conexión
sftp.close()
transport.close()
```
