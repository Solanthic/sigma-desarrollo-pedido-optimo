```python
import pandas as pd
import win32com.client as win32
import os
import time
from datetime import datetime
import os


df_alertas = pd.read_excel(
    'Resumen_carga.xlsx',
    sheet_name='resumen_pedido',
    header=4
)
df_alertas.to_csv('alertas_mail.csv', index=False)

import pandas as pd
from datetime import datetime
import os

df_alertas = pd.read_excel(
    'Resumen_carga.xlsx',
    sheet_name='resumen_pedido',
    header=4
)
df_filtrado = df_alertas[df_alertas['Con_algun_empuje'] == 1].copy()

df_filtrado['Fecha_ajuste'] = datetime.today().strftime('%Y%m%d')

columnas_finales = [
    'Fecha_ajuste',
    'Sap',
    'Sku',
    'Pedido_original_cadena',
    'Con_algun_empuje',
    'Piezas a cargar'
]
df_final = df_filtrado[columnas_finales]

archivo_concentrado = 'Concentrado_impulso.csv'

if os.path.exists(archivo_concentrado):
    df_existente = pd.read_csv(archivo_concentrado)
    df_concatenado = pd.concat([df_existente, df_final], ignore_index=True)
else:
    df_concatenado = df_final

df_concatenado.drop_duplicates(
    subset=['Sap', 'Sku', 'Pedido_original_cadena', 'Con_algun_empuje', 'Piezas a cargar'],
    inplace=True
)

df_concatenado.to_csv(archivo_concentrado, index=False)

df_filtrado = df_alertas[df_alertas['Con_algun_recorte'] == 1].copy()

df_filtrado['Fecha_ajuste'] = datetime.today().strftime('%Y%m%d')

columnas_finales = [
    'Fecha_ajuste',
    'Sap',
    'Sku',
    'Pedido_original_cadena',
    'Con_algun_recorte',
    'Piezas a cargar'
]
df_final = df_filtrado[columnas_finales]

archivo_concentrado = 'Concentrado_recorte.csv'

if os.path.exists(archivo_concentrado):
    df_existente = pd.read_csv(archivo_concentrado)
    df_existente = df_existente[columnas_finales] if all(col in df_existente.columns for col in columnas_finales) else df_existente
    df_concatenado = pd.concat([df_existente, df_final], ignore_index=True)
else:
    df_concatenado = df_final

df_concatenado.drop_duplicates(
    subset=['Sap', 'Sku', 'Pedido_original_cadena', 'Con_algun_recorte', 'Piezas a cargar'],
    inplace=True
)

df_concatenado.to_csv(archivo_concentrado, index=False)


#ruta_archivo = os.path.join(os.getcwd(), 'Envia_alertas_mail_test.xlsm')

#excel = win32.DispatchEx('Excel.Application')  
#excel.Visible = True  

#wb = excel.Workbooks.Open(ruta_archivo)

#wb.RefreshAll()

#excel.CalculateUntilAsyncQueriesDone()

#wb.Save()

#excel.Run("EnviarReportesPorGrupo")



```
